## library 불러오기

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import requests as re
from bs4 import BeautifulSoup as bs
import time
```

## 네이버 기사 Crawling

불러와야 할 주소  
https://search.naver.com/search.naver?&where=news&query=%EC%95%A0%ED%94%8C&sm=tab_pge&sort=0&photo=0&field=0&reporter_article=&pd=0&ds=&de=&docid=&nso=so:r,p:all,a:all&mynews=0&cluster_rank=32&start=11&refresh_start=0

selector: #sp_nws55 > dl > dt > a

```python
output = []
url = "https://search.naver.com/search.naver"
for rep in range(0,12):
    params = {
        'where':'news',
        'query':'애플',
        'sm':'tab_pge',
        'sort':'0',
        'photo':'0',
        'field':'0',
        'reporter_article':'',
        'pd':'0',
        'ds':'',
        'de':'',
        'docid':'',
        'nso':'so:r,p:all,a:all',
        'mynews':'0',
        'cluster_rank':'32',
        'start':str(rep*10+1)
    }
    res = re.get(url = url, params = params)
    soup = bs(res.text, 'html.parser')
    temp = soup.select('dl > dt > a')
    temp_output = [x.get("title") for x in temp]
    output.append(temp_output)
    print(rep)
    time.sleep(5)
```

```python
for i in output:
    print(i)
```

```python
# output의 형태, 12개의 list, 각 10개의 값
print(len(output), len(output[0]))
```

```python
# 하나의 list 안의 120개의 값으로 형태 변화
flat_list = []
for sublist in output:
    for item in sublist:
        flat_list.append(item)
```

## 잘못된 기사 제거하기

관심있는 '애플' 이외에 '애플망고','애플힙' 등의 값이 존재한다.  
이와 관련된 기사 제목은 제거하고자 한다.  
애플 양 옆의 단어가 무엇인 지 확인해보았다.

```python
a = 0
next_word = []
for i in range(len(flat_list)):
    if (flat_list[i].find('애플')+2) == (len(flat_list[i])):
        next_word.append(flat_list[i][flat_list[a].find('애플')-1]+'애플')
    else:
        next_word.append('애플'+flat_list[i][flat_list[a].find('애플')+2])
    a += 1
```

```python
next_word
```

```python
# 지워야 할 인덱스
index = []

# 애플망고 찾아내기
for i in range(len(next_word)):
    if next_word[i] == '애플망':
        index.append(i)
        
# 애플폭스 찾아내기
for i in range(len(next_word)):
    if next_word[i] == '애플폭':
        index.append(i)
        
# 애플수박 찾아내기
for i in range(len(next_word)):
    if next_word[i] == '애플수':
        index.append(i)
        
# 애플힙 찾아내기
for i in range(len(next_word)):
    if next_word[i] == '애플힙':
        index.append(i)
        
# 애플환불 찾아내기
for i in range(len(next_word)):
    if next_word[i] == '애플환':
        index.append(i)

print(index)
```

```python
index.sort(reverse=True)
for i in range(len(index)):
    flat_list.pop(index[i])
len(flat_list)
```
